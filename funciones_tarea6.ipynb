## 1. Calculadora de IMC (Índice de Masa Corporal)

```python
def calcular_imc(peso, altura):
   
    return peso / (altura ** 2)

peso = float(input("Ingrese su peso: "))
altura = float(input("Ingrese su altura: "))

imc = calcular_imc(peso, altura)

print(f"El IMC es: {imc:.2f}")
```

## IMC según los estándares de la OMS

```python
def calcular_imc(peso, altura):
   
    return peso / (altura ** 2)

def estándares_imc(imc):
    
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

peso = float(input("Ingrese su peso: "))
altura = float(input("Ingrese su altura: "))

imc = calcular_imc(peso, altura)

estándar = estándares_imc(imc)

print(f"El IMC es: {imc:.2f}")
print(f"Estándar: {estándar}")
```

## 2. Analizador de números

```python
def analizador_numeros(numero):
    resultado = {
        'Par_Impar': 'Par' if numero % 2 == 0 else 'Impar',
        'Positivo_Negativo_Cero': 'Positivo' if numero > 0 else 'Negativo' if numero < 0 else 'Cero',
        'Divisible_por_3': numero % 3 == 0,
        'Divisible_por_5': numero % 5 == 0,
        'Divisible_por_3_y_5': numero % 3 == 0 and numero % 5 == 0
    }
    return resultado

numero = int(input("Ingrese un número entero: "))
info = analizador_numeros(numero)

for clave, valor in info.items():
    print(f"{clave}: {valor}")
```
