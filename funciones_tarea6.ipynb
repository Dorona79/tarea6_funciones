## 1. Calculadora de IMC (Índice de Masa Corporal)

```python
def calcular_imc(peso, altura):
   
    return peso / (altura ** 2)

peso = float(input("Ingrese su peso: "))
altura = float(input("Ingrese su altura: "))

imc = calcular_imc(peso, altura)

print(f"El IMC es: {imc:.2f}")
```

## IMC según los estándares de la OMS

```python
def calcular_imc(peso, altura):
   
    return peso / (altura ** 2)

def estándares_imc(imc):
    
    if imc < 18.5:
        return "Bajo peso"
    elif 18.5 <= imc < 25:
        return "Peso normal"
    elif 25 <= imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

peso = float(input("Ingrese su peso: "))
altura = float(input("Ingrese su altura: "))

imc = calcular_imc(peso, altura)

estándar = estándares_imc(imc)

print(f"El IMC es: {imc:.2f}")
print(f"Estándar: {estándar}")
```

## 2. Analizador de números

```python
def analizador_numeros(numero):
    resultado = {
        'Par_Impar': 'Par' if numero % 2 == 0 else 'Impar',
        'Positivo_Negativo_Cero': 'Positivo' if numero > 0 else 'Negativo' if numero < 0 else 'Cero',
        'Divisible_por_3': numero % 3 == 0,
        'Divisible_por_5': numero % 5 == 0,
        'Divisible_por_3_y_5': numero % 3 == 0 and numero % 5 == 0
    }
    return resultado

numero = int(input("Ingrese un número entero: "))
info = analizador_numeros(numero)

for clave, valor in info.items():
    print(f"{clave}: {valor}")
```

## 3. Calculadora de área y perímetro de figuras geométricas

```python
import math

def calcular_figura(forma):
    if forma == "cuadrado":
        lado = float(input("Ingrese el lado del cuadrado: "))
        area = lado ** 2
        perimetro = 4 * lado
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    elif forma == "triángulo_equilátero":
        lado = float(input("Ingrese el lado del triángulo equilátero: "))
        area = (math.sqrt(3) / 4) * (lado ** 2)
        perimetro = 3 * lado
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    elif forma == "triángulo_rectángulo":
        base = float(input("Ingrese la base del triángulo rectángulo: "))
        altura = float(input("Ingrese la altura del triángulo rectángulo: "))
        area = (base * altura) / 2
        hipotenusa = math.sqrt(base ** 2 + altura ** 2)
        perimetro = base + altura + hipotenusa
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    elif forma == "triángulo_isósceles":
        lado_igual = float(input("Ingrese el lado igual del triángulo isósceles: "))
        base = float(input("Ingrese la base del triángulo isósceles: "))
        altura = math.sqrt(lado_igual ** 2 - (base / 2) ** 2)
        area = (base * altura) / 2
        perimetro = 2 * lado_igual + base
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    elif forma == "círculo":
        radio = float(input("Ingrese el radio del círculo: "))
        area = math.pi * radio ** 2
        perimetro = 2 * math.pi * radio
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    elif forma == "rectángulo":
        largo = float(input("Ingrese el largo del rectángulo: "))
        ancho = float(input("Ingrese el ancho del rectángulo: "))
        area = largo * ancho
        perimetro = 2 * (largo + ancho)
        return {"área": int(area) if area.is_integer() else area, "perímetro": int(perimetro) if perimetro.is_integer() else perimetro}
    
    else:
        return "Forma no soportada"

forma = input("Ingrese la forma geométrica (cuadrado, triángulo_equilátero, triángulo_rectángulo, triángulo_isósceles, círculo, rectángulo): ")
resultados = calcular_figura(forma)
print(f"Área: {resultados['área']} Perímetro: {resultados['perímetro']}")
```
